07 - **Faça um programa que receba dois números e mostre o maior. Se por acaso os dois números forem iguais, mostre a mensagem ''números iguais''.**

```python
x = int(input('Digite um número inteiro: \n'))
y = int(input('Digite outro número inteiro: \n'))

if x > y:
  print(f'O número {x} é maior que o número {y}.')
elif y > x:
  print(f'O número {y} é maior que o número {x}.')
else:
  print(f'Os números {x} e {y} são iguais.')
```
