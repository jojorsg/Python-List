13 - **Faça um algorítmo que calcule a média ponderada das notas de 3 provas. A primeira e a segunda prova tem peso 1 e a terceira tem peso 2. Ao final, mostrar a média do aluno e indicar se o aluno foi aprovado ou reprovado. A nota para aprovação deve ser igual ou superior a 60 pontos.**

```python
notap1 = float(input('Digite a nota da prova 1: '))
notap2 = float(input('Digite a nota da prova 2: '))
notap3 = float(input('Digite a nota da prova 3: '))

media = ((notap1 * 1) + (notap2 * 1) + (notap3 * 2)) / 4

if media >= 60:
  situacão = 'Aprovado'
else:
  situacão = 'Reprovado'

print(f'A situação do aluno é de {situacão}, com a Média de {media:.1f} pontos.')
```
