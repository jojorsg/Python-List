10 - **Faça um programa que receba a altura e o sexo de uma pessoa e calcule e mostre seu peso ideal, utilizando as seguintes fórmulas (onde h corresponde à altura).**
##### - Homens: (72.7 * h) - 58
##### - Mulheres: (62.1 * h) - 44.7

```python
altura = float(input("Digite sua altura (em metros): "))
sexo = input("Digite seu sexo (M para masculino, F para feminino): ")

if sexo == "M" or sexo == "m":
    peso_ideal = (72.7 * altura) - 58
    print(f'Seu peso ideal é {peso_ideal:.2f} kg.')
elif sexo == "F" or sexo == "f":
    peso_ideal = (62.1 * altura) - 44.7
    print(f'Seu peso ideal é {peso_ideal:.2f} kg.')
else:
    print("Sexo inválido. Somente M para masculino ou F para feminino.")
```
