06 - **Escreva um programa que, dados dois números inteiros, mostre na tela o maior deles, assim como a diferença entre ambos.**

```python
a = int(input('Digite um número inteiro: \n'))
b = int(input('Digite outro número inteiro: \n'))

if a > b:
  print(f'O número {a} é o maior dos dois com a diferença de {a - b} pontos.')
elif b > a:
  print(f'O número {b} é o maior dos dois com a diferença de {b - a} pontos.')
else:
  print(f'Não existe diferença e nem um número maior entre {a} e {b}.')
```
