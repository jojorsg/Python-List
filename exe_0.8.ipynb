08 - **Faça um programa que leia duas notas de um aluno, verifique se as notas são válidas e exiba a média dessas notas na tela. Uma nota válida deve ser, obrigatoriamente, um valor entre 0.0 e 10.0, onde caso a nota não possua um valor válido este, fato deve ser informado ao usuário e o programa termina.**

```python

nota1 = float(input('Digite a primeira nota: \n'))
nota2 = float(input('Digite a segunda nota: \n'))

if nota1 >= 0.0 and nota1 < 10.1 and nota2 >= 0.0 and nota2 < 10.1:
  result = nota1 + nota2
  media = result / 2
  print(f'A média será {media:.1f}.')
elif nota2 <= 0 or nota2 > 10:
  print('A segunda nota é inválida.')
else:
  print('A primeira nota não possue um valor válido.') 
```
