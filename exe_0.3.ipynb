03 - **Leia um número real. Se o número for positivo, imprima a raiz quadrada. Do contrário, imprima o número ao quadrado.**

```python
import math

x = float(input('Digite um número real: \n'))

if x > 0:
  raizx = math.sqrt(x)
  print(f'A raiz quadrada de {x} é {raizx:.2f}.')
elif x < 0:
  quadx = x**2
  print(f'Sendo {x} um número negativo, lhe daremos seu quadrado que é {quadx:.2f}.')
else:
  print(f'para {x}, tanto seu quadrado, quanto sua raiz quadrada é 0.')
```
