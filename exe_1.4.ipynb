14 - **A nota final de um estudante é calculada a partir de três notas atribuídas entre o intervalo de 0 até 10, respectivamente, a um trabalho de laboratório, a uma avaliação semestral e a um exame final. A média das três notas mencionadas anteriormente obedece aos pesos: Trabalho de Laboratório: 2; Avaliação Semestral: 3; Exame Final: 5. De acordo com o resultado, mostre na tela se o aluno está reprovado(média entre 0 e 2.9), de recuperação (entre 3 e 4.9) ou se foi aprovado. Faça todas as verificações necessárias.**

```python
nota1 = float(input('Nota do trabalho de laboratório: '))
nota2 = float(input('Nota da Avaliação Semestral: '))
nota3 = float(input('Nota do Exame Final: '))

media = ((nota1 * 2) + (nota2 * 3) + (nota3 * 5)) / 10

while nota1 < 0 or nota1 > 10 or nota2 < 0 or nota2 > 10 or nota3 < 0 or nota3 > 10:
  if nota1 < 0 or nota1 > 10:
    print(f'Digite uma nota válida para Trabalho de Laboratório.')
  elif nota2 < 0 or nota2 > 10:
    print(f'Digite uma nota válida para Avaliação Semestral.')
  elif nota3 < 0 or nota3 > 10:
    print(f'Digite uma nota válida para Exame Final.')
  nota1 = float(input('Nota do trabalho de laboratório: '))
  nota2 = float(input('Nota da Avaliação Semestral: '))
  nota3 = float(input('Nota do Exame Final: '))
  media = ((nota1 * 2) + (nota2 * 3) + (nota3 * 5)) / 10

if media >= 0 and media <= 2.9:
  print(f'O aluno está reprovado com a média {media:.1f}.')
elif media >= 3 and media <= 4.9:
  print(f'O aluno está de recuperação com a média {media:.1f}.')
else:
  print(f'O aluno está aprovado com a média {media:.1f}. PARABÉNS!!!')
```
