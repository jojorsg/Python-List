02 - **Leia um número fornecido pelo usuário. se esse número for positivo, calcule a raiz quadrada desse número. Se o número for negativo, mostre uma mensagem dizendo que o número é inválido.**

```python
import math

x = int(input('Digite um número: \n'))

if x > 0:
  raizx = math.sqrt(x)
  print(f'A raiz quadrada de {x} é {raizx:.2f}.')
elif x < 0:
  print('Esse número é invalido.')
else:
  print('A raiz quadrada de 0 é 0.')
```
