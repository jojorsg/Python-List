01 - **Faça um programa que receba dois números e mostre qual deles é o maior.**

```python
x = input('Digite um número: \n')
y = input('Digite outro número: \n')

if x > y: 
  print(f'O número {x} é maior que {y}.')
elif x < y:
  print(f'O número {y} é maior que {x}.')
else:
  print(f'Os números são iguais.')
  
```
