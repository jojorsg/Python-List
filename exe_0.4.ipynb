04 - **Faça um programa que leia um número e, caso ele seja positivo, calcule e mostre:**
##### . O número digitado ao quadrado.
##### . A raiz quadrada do número digitado.


```python
import math
x = int(input('Digite um número: \n'))

try:
  if x > 0:
    quadx = x**2
    raizx = math.sqrt(x)
    print(f'{x}² será {quadx:.2f}. \nA raiz quadrada de {x} é {raizx:.2f}.')
  elif x < 0:
    quadx = x**2
    print(f'o quadrado de {x} é {quadx:.2f}. Porém não existe raiz quadrada de números negativos.')
  else:
    print(f'Tanto o quadrado quanto a raiz quadrada de 0 é 0.')
except ValueError:
  print('Digite um número válido...')
```
