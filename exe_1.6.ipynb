16 - **Usando o switch, escreva um programa que leia um inteiro entre 1 e 12 e imprima o mês correspondente a esse número.**

```python
mes = int(input('Digite um número: '))

while mes < 1 or mes > 12:
  print(f'O número {mes} é inválido. Tente novamente...')
  mes = int(input('Digite um número: '))

def mes_do_ano(mes):
  while mes > 0 and mes < 13:
    if mes == 1:
      return 'Janeiro'
    elif mes == 2:
      return 'Fevereiro'
    elif mes == 3:
      return 'Março'
    elif mes == 4:
      return 'Abril'
    elif mes == 5:
      return 'Maio'
    elif mes == 6:
      return 'Junho'
    elif mes == 7:
      return 'Julho'
    elif mes == 8:
      return 'Agosto'
    elif mes == 9:
      return 'Setembro'
    elif mes == 10:
      return 'Outubro'
    elif mes == 11:
      return 'Novembro'
    elif mes == 12:
      return 'Dezembro'

print(f'O número {mes} corresponde ao mês de {mes_do_ano(mes)}.')
```
