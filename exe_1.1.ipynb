11 - **Escreva um programa que leia um número inteiro maior do que zero e devolva na tela a soma de todos os seus algorítimos. Por exemplo, ao número 251 corresponderá o valor 8 (2 + 5 + 1). Se o número lido não for maior do que zero, o programa terminá com a mensagem "Número inválido".**

```python
num = input("Digite um número inteiro maior do que zero: ")

if int(num) > 0:
  soma = sum(int(digito) for digito in num)
  print(f'A soma dos dígitos é {soma}.')
else:
  print('Número inválido.')
```
