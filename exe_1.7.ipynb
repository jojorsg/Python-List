17 - **Faça um programa que calcule e mostre a área de um trapézio. Sabe-se que: a = (basemaior + basemenor) * altura / 2. Lembre-se que a basemaior e a basemenor devem ser números maiores que zero.**

```python
basemaior = float(input('Digite um número para a base maior: '))
basemenor = float(input('Digite um número para a base menor: '))
altura = float(input('Digite o valor da altura: '))

while basemaior <= 0:
  print('Número inválido. A base maior não pode ser menor que zero.')
  pergunta = int(input('Digite 1 para tentar novamente ou 2 para encerrar o programa: '))
  if pergunta == 1:
    basemaior = float(input('Digite um número para a base maior: '))
  elif pergunta == 2:
    print('Programa encerrado. Obrigado!')
    break
  else:
    print('Número inválido.')
  basemaior = float(input('Digite um número para a base maior: '))

while basemenor <= 0:
  print('Número inválido. A base menor não pode ser menor que zero.')
  pergunta = int(input('Digite 1 para tentar novamente ou 2 para encerrar o programa: '))
  if pergunta == 1:
    basemenor = float(input('Digite um número para a base menor: '))
  elif pergunta == 2:
    print('Programa encerrado. Obrigado!')
    break
  else:
    print('Número inválido.')
  basemenor = float(input('Digite um número para a base menor: '))

while altura == 0:
  print('Qualquer número multiplicado por zero é zero.')
  pergunta = int(input('Digite 1 para tentar novamente ou 2 para encerrar o programa: '))
  if pergunta == 1:
    altura = float(input('Digite o valor da altura: '))
  elif pergunta == 2:
    print('Programa encerrado. Obrigado!')
    break
  else:
    print('Número inválido.')
    
if altura != 0 and basemaior > 0 and basemenor > 0:
  area = (basemaior + basemenor) * altura / 2
  print(f'A área do trapezio é de {area}m².')
```
