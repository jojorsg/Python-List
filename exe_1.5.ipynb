15 - **Usando switch, escreva um programa que leia um inteiro entre 1 e 7 e imprima o dia da semana correspondente a este número. Isto é, domingo se 1, segunda-feira se 2, e assim por diante.**

```python
x = int(input('Digite um número inteiro entre 1 e 7: \n'))
def dia_da_semana(x):
    if x == 1:
        return "domingo"
    elif x == 2:
        return "segunda-feira"
    elif x == 3:
        return "terça-feira"
    elif x == 4:
        return "quarta-feira"
    elif x == 5:
        return "quinta-feira"
    elif x == 6:
        return "sexta-feira"
    elif x == 7:
        return "sábado"
    else:
        return "Número inválido"

print(f"O dia da semana correspondente ao número {x} é {dia_da_semana(x)}.")
```
