19 - **Faça um programa para verificar se um determinado número inteiro é divisível por 3 ou 5, mas não simultaneamente pelos dois.**

```python
num = int(input('Digite um número inteiro: '))

if num % 3 == 0:
  print(f'O número {num} é divisível por 3.')
elif num % 5 == 0:
  print(f'O número {num} é divisível por 5.')
elif num % 3 == 0 and num % 5 == 0:
  print(f'O número {num} sendo divisível por 3 e por 5 ao mesmo tempo torna-se inválido.')
else:
  print(f'O número {num} não é divisível por 3 e não é divisível por 5')
```
