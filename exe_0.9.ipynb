09 - **Leia o salário de um trabalhador e o valor de uma prestação de um empréstimo. Se a prestação for maior que 20% do salário imprima: "empréstimo não concedido", caso contrário imprima: "empréstimo concedido".**

```python
x = float(input('Digite o valor do salário: \n'))
y = float(input('Digite o valor da prestação: \n'))
sal = x * (20 / 100)
print(f'{sal:.2f}')

if y > sal:
  print(f'Empréstimo não concedido')
else:
  print(f'Empréstimo concedido')
```
