12 - **Ler um número inteiro. Se o número lido for negativo, escreva a mensagem "Número inválido". Se o número for positivo, Calcular o algorítmo deste número.**

```python
num = int(input('Digite um número inteiro: \n'))

if num >= 0:
  soma = sum(int(digito) for digito in str(num))
  print(f'O calculo do algorítmo do número {num} é {soma}.')
else:
  print('Número inválido')
```
